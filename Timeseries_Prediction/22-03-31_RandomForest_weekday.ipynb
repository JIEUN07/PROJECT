````python
# 새로운 feature처리 함수

def lamp_timestamps(data_1, data_2):
  # 공항별 departure_runway, arrival_runway, train_label 쪼갠 후 처리
  for j in range(len(airport_directories)):
    de = data_1.loc[data_1["airport"] == airport_directories[j].name]
    de["timestamp"] = pd.to_datetime(de["timestamp"], infer_datetime_format=True)

    ar = data_2.loc[data_2["airport"] == airport_directories[j].name]
    ar["timestamp"] = pd.to_datetime(ar["timestamp"], infer_datetime_format=True)

    train = train_labels.loc[train_labels["airport"] == airport_directories[j].name].reset_index()

    for i in range(len(train)):
      # lookahead데이터 변수 설정
      a = train["lookahead"][i]

      # departure_runway 데이터 처리
      data_1_unique = de.loc[(de["timestamp"] >= (train["timestamp"][i] - timedelta(hours=4, ))) & (de["timestamp"] < (train["timestamp"][i] + timedelta(minutes= a, )))]
      data_1_unique = data_1_unique["departure_runway"].unique()
      data_1_unique.sort()
      data_1_str = "_".join(data_1_unique)
      train.loc[(train["timestamp"] == train["timestamp"][i]) & (train["lookahead"] == a), "D"] = data_1_str

      # arrival_runway 데이터 처리
      data_2_unique = ar.loc[(ar["timestamp"] >= (train["timestamp"][i] - timedelta(hours=4, ))) & (ar["timestamp"] < (train["timestamp"][i] + timedelta(minutes= a, )))]
      data_2_unique = data_2_unique["arrival_runway"].unique()
      data_2_unique.sort()
      data_2_str = "_".join(data_2_unique)
      train.loc[(train["timestamp"] == train["timestamp"][i]) & (train["lookahead"] == a), "A"] = data_2_str

    # 두번째 반복부터 concat (첫번째 반복에는 train_all이 없어 concat이 안됨)
    if j != 0:
      train_all = pd.concat([train_all, train])
    else:
      train_all = train
  
  return train_all
````

````python
# 새로운 feature처리 함수 (unique 개수제한)

def timestamps_unique(data_1, data_2):
  # 공항별 departure_runway, arrival_runway, train_label 쪼갠 후 처리
  for j in range(len(airport_directories)):
    de = data_1.loc[data_1["airport"] == airport_directories[j].name]
    de["timestamp"] = pd.to_datetime(de["timestamp"], infer_datetime_format=True)

    ar = data_2.loc[data_2["airport"] == airport_directories[j].name]
    ar["timestamp"] = pd.to_datetime(ar["timestamp"], infer_datetime_format=True)

    train = train_labels.loc[train_labels["airport"] == airport_directories[j].name].reset_index()

    for i in range(len(train)):
      # lookahead데이터 변수 설정
      a = train["lookahead"][i]

      # departure_runway 데이터 처리
      data_1_unique = de.loc[(de["timestamp"] >= (train["timestamp"][i] - timedelta(hours=4, ))) & (de["timestamp"] < (train["timestamp"][i] + timedelta(minutes= a, )))]
      data_1_unique = data_1_unique["departure_runway"].unique()
      data_1_unique.sort()
      data_1_unique= data_1_unique[:3]
      data_1_str = "_".join(data_1_unique)
      train.loc[(train["timestamp"] == train["timestamp"][i]) & (train["lookahead"] == a), "D"] = data_1_str

      # arrival_runway 데이터 처리
      data_2_unique = ar.loc[(ar["timestamp"] >= (train["timestamp"][i] - timedelta(hours=4, ))) & (ar["timestamp"] < (train["timestamp"][i] + timedelta(minutes= a, )))]
      data_2_unique = data_2_unique["arrival_runway"].unique()
      data_2_unique.sort()
      data_2_unique = data_2_unique[:3]
      data_2_str = "_".join(data_2_unique)
      train.loc[(train["timestamp"] == train["timestamp"][i]) & (train["lookahead"] == a), "A"] = data_2_str

    # 두번째 반복부터 concat (첫번째 반복에는 train_all이 없어 concat이 안됨)
    if j != 0:
      train_all = pd.concat([train_all, train])
    else:
      train_all = train
  
  return train_all
````

````python
# 가중치 함수

def final_weight(sub_format):
  for j in range(len(unique_rate)):
    a = sub_format.loc[sub_format['airport'] == airport_directories[j].name].reset_index(drop=True)
    b = submission_0.loc[submission_0['airport'] == airport_directories[j].name].reset_index(drop=True)

    for i in tqdm(range(len(b))):
      a.loc[(a['timestamp'] == b['timestamp'][i]) & (a['lookahead'] == b['lookahead'][i]) & (a['config'] == b['pre_config'][i]), 'active'] = b['active'][i]

    if j != 0:
      all = pd.concat([all, a])
    else:
      all = a

  return all.reset_index(drop=True)
````
